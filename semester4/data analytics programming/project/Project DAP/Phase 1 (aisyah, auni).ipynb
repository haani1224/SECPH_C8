```python
import pandas as pd
```

```python
# Load the datasets
fuel_df = pd.read_csv('FuelConsumption.csv')
cars_df = pd.read_csv('cars_2025.csv')

# Standardize column names to lowercase and strip spaces
fuel_df.columns = fuel_df.columns.str.lower().str.strip()
cars_df.columns = cars_df.columns.str.lower().str.strip()

# Strip whitespace and convert all values in the 'make' and 'model' columns to uppercase.
fuel_df['make'] = fuel_df['make'].str.strip().str.upper()
fuel_df['model'] = fuel_df['model'].str.strip().str.upper()
cars_df['maker'] = cars_df['maker'].str.strip().str.upper()
cars_df['model'] = cars_df['model'].str.strip().str.upper()

# Check for missing values
print("Missing values in FuelConsumption.csv:")
print(fuel_df.isnull().sum())
print("\nMissing values in Cars2025.csv:")
print(cars_df.isnull().sum())
print()

# Check for duplicate rows
print("Number of duplicate rows in FuelConsumption.csv:", fuel_df.duplicated().sum())
print("Number of duplicate rows in Cars2025.csv:", cars_df.duplicated().sum())

# Remove duplicates
cars_df = cars_df.drop_duplicates()
```

```python
# Check the cleaned datasets shape
print("Cleaned FuelConsumption.csv shape:", fuel_df.shape)
print("Cleaned cars_2025.csv shape:", cars_df.shape)
print()

# Preview the cleaned data
print(fuel_df.head())
print(cars_df.head())
```

```python
# Save cleaned versions to new CSV files
fuel_df.to_csv('FuelConsumption_cleaned.csv', index=False)
cars_df.to_csv('Cars2025_cleaned.csv', index=False)

print("Cleaned datasets saved as 'FuelConsumption_cleaned.csv' and 'Cars2025_cleaned.csv'")
```
